```python
import pandas as pd
```

```python
df = pd.read_excel('AstroTable.xlsx')
result = {}
for idx, row in df.iterrows():
    name = row.iloc[0]          
    attributes = row.iloc[1:13]
    result[name] = attributes.tolist()
```

```python
for name, attributes in result.items():
    attributes_str = str(attributes).replace("'", '"')
    print(f'"{name}": {attributes_str},')
```
